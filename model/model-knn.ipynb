```python
# IMPORT LIBRARY
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
import joblib
```

```python
# IMPORT DATA
dataset = pd.read_csv("../dataset/datasetmerged_duplicate7x.csv")
dataset
```

```python
#LABEL & FITUR
label = dataset['Title']
fitur = ['BodyPart','Type', 'Rating']
fitur = dataset[fitur]
fitur
```

```python
label
```

```python
features_encoded = pd.get_dummies(fitur, columns=['BodyPart', 'Type'])
features_encoded
```

```python
label_encoder = LabelEncoder()
labels_encoded = label_encoder.fit_transform(label)
labels_encoded = labels_encoded.reshape(-1, 1)
labels_encoded
```

```python
features_encoded = features_encoded.to_numpy()
features_encoded = features_encoded.astype(float)
features_encoded
```

```python
print(features_encoded.shape)
print(labels_encoded.shape)
```

```python
# BAGI DATA UJI DAN DATA LATIH
X_train, X_test, y_train, y_test = train_test_split(features_encoded, labels_encoded, test_size=0.2, random_state=42)
```

```python
# MODEL KNN
model = KNeighborsClassifier(n_neighbors=3)
model.fit(X_train, y_train)
```

```python
# PREDIKSI DENGAN DATA UJI
predictions = model.predict(X_test)
```

```python
#EVALUASI MODEL
accuracy = model.score(X_test, y_test)
print(f"Akurasi model: {accuracy}")
```
