```python
# IMPORT LIBRARY
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
import tensorflow as tf
import keras
```

```python
# IMPORT DATA
dataset = pd.read_csv("../dataset/datasetFIKSS.csv")
dataset
```

```python
#LABEL & FITUR
label = dataset['Title']
fitur = ['BodyPart','Type', 'Rating']
fitur = dataset[fitur]
fitur
```

```python
label
```

```python
features_encoded = pd.get_dummies(fitur, columns=['BodyPart', 'Type'])
features_encoded
```

```python
label_encoder = LabelEncoder()
labels_encoded = label_encoder.fit_transform(label)
labels_encoded
```

```python
features_encoded = features_encoded.to_numpy()
features_encoded = features_encoded.astype(float)
features_encoded
```

```python
print(features_encoded.shape)
print(labels_encoded.shape)
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Dense(128, input_dim=features_encoded.shape[1], activation='relu'),
    tf.keras.layers.Dense(512, activation='relu'),
    tf.keras.layers.Dense(1024, activation='relu'),
    tf.keras.layers.Dense(512, activation='relu'),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dense(354, activation='softmax')
])
model.summary()
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.fit(features_encoded, labels_encoded, epochs=50, batch_size=32)
```
